# Code Pemograman (Python 3)

untuk menggunakan code, perlu mengimport modul numpy, matplotlib dan math

```python
import numpy as np
import matplotlib.pyplot as plot
import math
```

dan untuk membuat gelombang, digunakan class sebagai bentuk penerapan *Object Oriented Programming (OOP)*

```python
class gelombang:
    
    def __init__(self, amplitudo, frekuensi, waktu=1,frekuensisampling=100, fasa=0):
        self.amplitudo          = float(amplitudo)
        self.frekuensi          = float(frekuensi)
        self.fasa               = float(fasa)
        self.frekuensisampling  = float(frekuensisampling)
        self.waktu              = np.arange(0,waktu+1/frekuensisampling,1/frekuensisampling)
    
    def y(self,trigonometri):
        jenisgelombang={'sin':np.sin,'cos':np.cos}
        return [x*self.amplitudo for x in jenisgelombang[trigonometri](
            [(2*math.pi*self.frekuensi*x)+self.fasa for x in self.waktu]
        )]
```

class ini dengan method `y(t)`. Selanjutnya akan dideklarasikan sebuah variabel dengan object gelombang tersebut, dimana

`gelombang(Amplitudo, Frekuensi, Waktu, Frekuensi Sampling, Fasa)`

```python
gelombang = gelombang(1,1,1,10)
```

apabila object ini mengambil metode *y(t)*, akan mengembalikan list dalam bentuk float dan jenis gelombang t (*sin* atau *cos*) dimana t adalah string

```python
gelombang.y('sin')
```

list ini dapat diinterpretasikan dalam 3 bentuk grafik matplotlib dengan function

```python
def plotter(gelombang):
    def plotproperties():
        plot.xlabel('Waktu(s)')
        plot.ylabel('Amplitudo(m)')
        plot.grid(True, which='both',color='gainsboro')
        plot.axhline(y=0, color='k')
        plot.ylim([-(gelombang.amplitudo+3),(gelombang.amplitudo+3)])
    
    def plot1():
        plot.plot(gelombang.waktu, gelombang.y('sin'))

    def plot2():
        (markers, stemlines, baseline) = plot.stem(
            gelombang.waktu, 
            gelombang.y('sin'),
            linefmt='gainsboro',
            basefmt=" ",
            markerfmt='D'
        )
        plot.setp(markers,
            markersize=3,
            markerfacecolor='orange',
            markeredgecolor='orange')
    
    plot.subplot(1, 3, 1)
    plotproperties()
    plot1()

    plot.subplot(1, 3, 2)
    plotproperties()
    plot.title('Gelombang A')
    plot2()

    plot.subplot(1, 3, 3)
    plotproperties()
    plot1()
    plot2()
```

kemudian, untuk menampilkan grafik dari gelombang tersebut dalam bentuk figur 30 inci x 8 inci dengan dpi 200 adalah

```python
fig=plot.figure(figsize=(30,8), dpi= 200, facecolor='w', edgecolor='k')
plotter(gelombang(6,3,1,75))
plot.show()
```
